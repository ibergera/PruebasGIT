```python
%config IPCompleter.greedy=True
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv(r'C:\Users\ibergera\Desktop\CS DATOS\Arg X La Edu\csv_files_app_alumno.csv')
df
```

```python
df2=df.dropna()
filtro3 =(df2['sexo'] != 'nc') & (df2['indice_socioeconomico'] != 'nc') & (df2['nivel_desemp_matematica'] != 'nc') & (df2['nivel_desemp_lengua'] != 'nc') & (df2['nivel_desemp_ciencias_sociales'] != 'nc' ) & (df2['nivel_desemp_ciencias_naturales'] != 'nc') & (df2['repeticion_primaria'] != 'nc') & (df2['repeticion_secundaria'] != 'nc')
df2=df2[filtro3]
```

```python
df2
```

```python
df5= df2[['escuela_id']]
df5
```

```python
sexo_df2_test= pd.get_dummies(df2['sexo'])
ise_df2_test= pd.get_dummies(df2['indice_socioeconomico'])
repetir_df2_test= pd.get_dummies(df2['repeticion_primaria'])
repetirsecu_df2_test= pd.get_dummies(df2['repeticion_secundaria'])

df5=pd.concat([df5,ise_df2_test],axis=1)
df5=pd.concat([df5,repetir_df2_test],axis=1)
df5=pd.concat([df5,repetirsecu_df2_test],axis=1)
df5=pd.concat([df5,sexo_df2_test],axis=1)
```

```python
df5.columns
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_ml=scaler.fit_transform(df5)
```

```python
from sklearn.decomposition import PCA
pca=PCA(5) 
pca.fit(df_ml)
X_pca=pca.transform(df_ml)
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.show()
```

```python
target = 'nivel_desemp_matematica'
X= X_pca
y= df2[target]

# qué porciento del dataset usaremos para entrenar el modelo?
n_train = 0.8
num_train = int(n_train*X.shape[0])
# train
X_train = X[:num_train]
y_train = y[:num_train]
# test
X_test = X[num_train:]
y_test = y[num_train:]
 
```

```python
from sklearn.neighbors import KNeighborsClassifier
clasificador = KNeighborsClassifier(n_neighbors=30)
clasificador.fit(X_train, y_train)
prediccion = clasificador.predict(X_test)
```

```python
# métricas de evaluación de modelos
from sklearn.metrics import accuracy_score

 

print('Evaluación del ml:')
print('Test: ', accuracy_score(y_test, prediccion))
print('-----------------------------------------')
```
